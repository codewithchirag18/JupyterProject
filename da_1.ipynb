# In thus file we are using "Used CAR PRICING" DATA SET

```python
import pandas as pd
import numpy as np
```

# Reading file

```python
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data")
```

# Storing the name of the columns in the variable " header"

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

# Assigning all the column name from the variable "headers"

```python
df.columns = headers
```


# Printing the Data Frame "df"

```python
df
```

# Replacing the "?" with Nan value using "np.nan"

```python
df1=df.replace('?',np.nan)
```

```python
df1
```

# Dropping all the row and columns which store the value Nan

```python
df= df1.dropna(subset =["price"],axis=0)
```

```python
df.loc[:,"price"]=df["price"].astype(int)
```

```python
df.dtypes
```

# BINNING TOPIC - Converting numerical value into categorical values

```python
bins=np.linspace(df["price"].min(),df["price"].max(),4)
```

```python
group_names=["low","medium","high"]
```

```python
df=df.copy()
df["price_binned"]=pd.cut(df["price"],bins,labels=group_names,include_lowest=True)
```

```python
df["price_binned"].head()
```

```python
df["price_binned"].value_counts()
```

# Using MATPLOTLIB.PYPLOT - to visualize the categorical variables

```python
import matplotlib.pyplot as plt
```

```python
df["price_binned"].value_counts().plot(kind="bar")
plt.xlabel("Price Category")
plt.ylabel("count")
plt.title("nummber of cars in each price category")
plt.show()
```
