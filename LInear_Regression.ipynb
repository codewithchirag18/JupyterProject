# MODULE - Model Development

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# _** TOPIC - LINEAR REGRESSION**_


* **Importing linear module from scikit-leanr**

```python
from sklearn.linear_model import LinearRegression
```

# Reading the file

```python
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data")
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns=headers
```

# Replacing all "?" value with Nan

```python
df.replace("?",np.nan,inplace=True)
```

# Dropping all the Nan values

```python
df.dropna(inplace=True)
```

# Changing the data type of price from 'object' to 'int'

```python
df["price"]=df["price"].astype(int)
```

# Checking the data type of price using .info()

```python
df['price'].info()
```

# Simple linear regression

# Creating the Model
# lm is a variable(short for linear model)
# LinearRegression() creates an empty regression model which can train(fit) data

```python
lm = LinearRegression()
```

# Define the predictor variable and target variable

```python
x=df[["highway-mpg"]]
y=df["price"]
```

# Fitiing the Model (we can say training the model)
* fit() -: Train the Model
* x=independent variable , y = independent vaiable
* it calculates the slope(m,b1) and intercept(c,b0)

```python
lm.fit(x,y)
```

# Checking the slope

```python
lm.coef_
```

# Checking the intercept

```python
lm.intercept_
```

# Making Prediction
* predict -:uses line we fitted to predict value of Y from X

```python
Yhat=lm.predict([[20]])
```

```python
Yhat
```
